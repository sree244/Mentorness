**Decision Tree classifier without using limitimg criteria**

```python
X=df8.drop('SALARY',axis=1)
X
```

```python
Y=df8['SALARY']
Y
```

```python
len(X)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,random_state=10)

from sklearn.tree import DecisionTreeClassifier
modeldtc1=DecisionTreeClassifier()

modeldtc1.fit(xtrain,ytrain)
print(modeldtc1.score(xtrain,ytrain))
print(modeldtc1.score(xtest,ytest))
```

```python
train_accuracy=[]
test_accuracy=[]
for i in range(100):
  xtrain,xtest,ytrain,ytest=train_test_split(X,Y,random_state=i)
  m1=DecisionTreeClassifier()
  m1.fit(xtrain,ytrain)
  train_accuracy.append(m1.score(xtrain,ytrain))
  test_accuracy.append(m1.score(xtest,ytest))
```

```python
import matplotlib.pyplot as plt
plt.plot(train_accuracy)
plt.plot(test_accuracy)
plt.show()
```

```python
print(max(test_accuracy))
print(test_accuracy.index(max(test_accuracy)))
```

**Tune limiting criteria of Decision Tree Classifier to avoid overfitting**

```python
m2=DecisionTreeClassifier(max_depth=3)
m2.fit(xtrain,ytrain)
print(m2.score(xtrain,ytrain))
print(m2.score(xtest,ytest))
print("---------------------------------------")
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus

dot_data = export_graphviz(m2,feature_names=xtrain.columns)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
modeld2=DecisionTreeClassifier(max_depth=4,min_samples_leaf=40)
modeld2.fit(xtrain,ytrain)

print(modeld2.score(xtrain,ytrain))
print(modeld2.score(xtest,ytest))
print('--------------------------------')

dot_data = export_graphviz(modeld2,feature_names=xtrain.columns)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(ytrain,modeld2.predict(xtrain)))
print('-----------------------------------')
print(confusion_matrix(ytest,modeld2.predict(xtest)))
```

**Oversampling the data to fix biased data**

```python
from imblearn.over_sampling import RandomOverSampler
res=RandomOverSampler()

Xres,Yres=res.fit_resample(X,Y)

from sklearn.model_selection import train_test_split
xtrainS,xtestS,ytrainS,ytestS=train_test_split(Xres,Yres,random_state=100)

from sklearn.tree import DecisionTreeClassifier
modeld3=DecisionTreeClassifier()

modeld3.fit(xtrain,ytrain)

print(modeld3.score(xtrainS,ytrainS))
print(modeld3.score(xtestS,ytestS))
```

**Random forest classifier**

```python
xtrain,xtest,ytrain,ytest=train_test_split(Xres,Yres,random_state=90)

# Scaling of Data
from sklearn.preprocessing import MinMaxScaler
scl=MinMaxScaler()

scl.fit(xtrain)               # scale down all the features value between 0 to 1

xtrainS = scl.transform(xtrain)
xtestS = scl.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
rmodel = RandomForestClassifier(max_depth=11,random_state=3,min_samples_split=7)
rmodel1 = RandomForestClassifier(min_samples_leaf=20)
rmodel2 = RandomForestClassifier(n_estimators=40,min_samples_leaf=50)  # no.of estimators means no.of trees

rmodel.fit(xtrainS,ytrain)
rmodel1.fit(xtrainS,ytrain)
rmodel2.fit(xtrainS,ytrain)
```

```python
print(rmodel.score(xtrainS,ytrain))
print(rmodel.score(xtestS,ytest))
print('------------------------')
print(rmodel1.score(xtrainS,ytrain))
print(rmodel1.score(xtestS,ytest))
print('------------------------')
print(rmodel2.score(xtrainS,ytrain))
print(rmodel2.score(xtestS,ytest))
```

```python
train_acc=[]
test_acc=[]

for k in range(2,70):
  rmodel = RandomForestClassifier(max_depth=11,random_state=3,min_samples_split=k)

  rmodel.fit(xtrainS,ytrain)

  train_acc.append(rmodel.score(xtrainS,ytrain))
  test_acc.append(rmodel.score(xtestS,ytest))
```
